## Web Scrappig FBREF project in progresss


```python
import pandas as pd
```

```python
url = 'https://fbref.com/en/squads/822bd0ba/Liverpool-Stats'
```

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests
```

```python
data  = requests.get(url).text 
soup = BeautifulSoup(data,"html5lib")
```

```python
print(soup.prettify()) 
```

```python
table = soup.find('div', attrs = {'class' : 'table_container tabbed current', 'id':'div_stats_passing_9'})  
```

```python
print(table.prettify())
```

```python
tab = table.find('tr', attrs = {'class' : 'over_header'})  
```

```python
print(tab.prettify())
```

```python
table_headers = []
tab1 = table.find('thead').find_all('tr')
for names in tab1[1].find_all('th'):
    table_headers.append(names.text)

for i, names in enumerate(table_headers):
    print(str(i) + "____" + names)
```

```python
table_headers[2] = 'Position'   
table_headers[5] = 'TotalCmp'
table_headers[6] = 'TotalAtt'
table_headers[7] = 'TotalCmp%'
table_headers[9] = 'TotalPrgDist'
#short 
table_headers[10] = 'ShortCmp'
table_headers[11] = 'ShortAtt'
table_headers[12] = 'ShortCmp%'

#medium
table_headers[13] = 'MediumCmp'
table_headers[14] = 'MediumAtt'
table_headers[15] = 'MediumCmp%'

#long
table_headers[16] = 'LongCmp'
table_headers[17] = 'LongAtt'
table_headers[18] = 'LongCmp%'


print(table_headers)
```

```python
passing_data  = pd.DataFrame(columns=table_headers)
passing_data.info()
```

```python

players_info = table.find('tbody').find_all('tr')
for data in players_info:
    full_name = data.find_all("th")
    details= data.find_all("td")
    player = full_name[0].text
    nation = details[0].text
    position = details[1].text      
    age = details[2].text 
    ninety90 = details[3].text 
    TotalCmp = details[4].text 
    TotalAtt = details[5].text 
    TotalCmpPer = details[6].text 
    ToDist = details[7].text 
    TotalPrgDist  = details[8].text 
    ShortCmp  = details[9].text 
    ShortAtt = details[10].text 
    ShortCmpPer = details[11].text 
    MediumCmp = details[12].text 
    MediumAtt  = details[13].text
    MediumCmpPer = details[14].text 
    LongCmp  = details[15].text 
    LongAtt = details[16].text 
    LongCmpPer = details[17].text 
    Ast = details[18].text 
    xAG = details[19].text 
    xA = details[20].text 
    AxAG = details[21].text 
    KP= details[22].text 
    onethird = details[23].text 
    PPA = details[24].text 
    CrsPA = details[25].text 
    PrgP = details[26].text 
    Matches = details[27].text
    passing_data = passing_data.append({'Player': player,'Nation':nation,'Position':position, 'Age':age, '90s':ninety90,
    'TotalCmp':TotalCmp, 'TotalAtt':TotalAtt, 'TotalCmp%': TotalCmpPer,'TotDist':ToDist,'TotalPrgDist':TotalPrgDist,
    'ShortCmp':ShortCmp,'ShortAtt':ShortAtt, 'ShortCmp%':ShortCmpPer,'MediumCmp':MediumCmp,'MediumAtt':MediumAtt,
    'MediumCmp%':MediumCmpPer,'LongCmp':LongCmp, 'LongAtt':LongAtt,'LongCmp%':LongCmpPer,'Ast':Ast,'xAG':xAG,
    'xA':xA, 'A-xAG':AxAG,'KP':KP,'1/3':onethird,'PPA':PPA,'CrsPA':CrsPA,'PrgP':PrgP,'Matches':Matches},ignore_index=True)

    
```
