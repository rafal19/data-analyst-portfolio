```python
import pandas as pd 
```

```python
data = pd.read_csv("World Universites Ranking 2023.csv",encoding = 'unicode_escape')
```

```python
#I have removed manually rows contain words "Explore", "Not accredited" in Excel"
df = pd.DataFrame(data)
df.head(5)
```

```python
# I have dropped the first record from the dataset
df.drop([0],axis=0,inplace=True)
```

```python
#I have extracted names from even records
lst = []
for index in df.index:
    if index %2 == 0:
        lst.append(df["Name"][index])
```

```python
# Checking numbers of records list should contain half less records than origin data
print(len(df))
print(len(lst)) 
df.head(5)
```

```python
# I have removed useless records by dropping NA values
df = df.dropna(subset=['Rank'])
print(len(df))
```

```python
#I have created new column 'Country and assigned list of extracted countries '
df.insert(6,'Country', lst)
df.head(5)
```

```python
#Resolving of wrong index by iterating over len of dataset
indexx = [f for f in range(len(df))]
idx = pd.Series(indexx)
df.set_index(idx)
#Saving data to csv
df.to_csv("universities.csv", index=False)
```

```python
data = pd.read_csv('universities.csv')
```

```python
#Partly cleaned data 
data.head(5)
```

```python
#Unifying Female:Male ratio, range of cases across this column is covered by two funtions below, last part of cleaning

def ratio_fractual(column):
    female = round(float(column),2)*100
    population = female + 100
    ratio_women = int(round((female/population)*100,0))
    ratio_men = 100 - ratio_women
    output = str(ratio_women) + ":" + str(ratio_men)
    return output
        
def ratio_cleaning(column):
    column = str(column)
    if len(column)>9 or len(column) == 4 or (len(column) == 7 and column[0] == '0'):
        column = ratio_fractual(column)
    if len(column) == 8 or len(column) ==9:
        column = column[0:5]
    if len(column) == 7:
        column = column.replace(" ","")
    if column == 'nan':
        column = ""
    return column
    
data['Female:Male Ratio']= data['Female:Male Ratio'].apply(ratio_cleaning)

data.head(5)
```

```python
data.to_csv("universities.csv", index=False, encoding="utf-32" )
```
